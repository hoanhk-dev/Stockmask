```python
import yfinance as yf

ticker = yf.Ticker("1812.T")  # Kajima Corporation

# ===== 1. BALANCE SHEET =====
bs = ticker.balance_sheet

# Lấy năm mới nhất
total_assets = bs.loc["Total Assets"].iloc[0]
total_liabilities = bs.loc["Total Liabilities Net Minority Interest"].iloc[0]

# Book Value of Assets (theo công thức bạn chọn)
book_value_assets = total_assets - total_liabilities


# ===== 2. MARKET VALUE =====
info = ticker.info

share_price = info.get("currentPrice")
shares_outstanding = info.get("sharesOutstanding")

market_value_assets = share_price * shares_outstanding


# ===== 3. ASSET REVALUATION RESERVE (PROXY) =====
asset_revaluation_reserve = market_value_assets - book_value_assets


# ===== 4. PRINT =====
print("=== Asset Revaluation Reserve (Market-based Proxy) ===")
print(f"Total Assets: {total_assets:,.0f}")
print(f"Total Liabilities: {total_liabilities:,.0f}")
print(f"Book Value of Assets: {book_value_assets:,.0f}")
print(f"Market Value (Equity): {market_value_assets:,.0f}")
print(f"Asset Revaluation Reserve (Proxy): {asset_revaluation_reserve:,.0f}")

```

```python
import yfinance as yf
import pandas as pd

ticker = yf.Ticker("1812.T")

# --- Income statement ---
income = ticker.financials
income_alt = ticker.income_stmt

# EBIT (fallback to Operating Income)
if "EBIT" in income.index:
    EBIT = income.loc["EBIT"]
else:
    EBIT = income_alt.loc["Operating Income"]

# Tax rate (single value, apply across years)
tax_rate = ticker.financials.loc["Tax Rate For Calcs"].iloc[0]

# --- Balance sheet ---
bs = ticker.balance_sheet

fixed_assets = bs.loc["Net PPE"]
current_assets = bs.loc["Current Assets"]
current_liabilities = bs.loc["Current Liabilities"]
goodwill = bs.loc["Goodwill And Other Intangible Assets"]

# --- Calculations ---
NOPAT = EBIT * (1 - tax_rate)

NWC = current_assets - current_liabilities

Invested_Capital = fixed_assets + NWC + goodwill

ROIC = (NOPAT / Invested_Capital).sort_index()

# --- Output ---
result = pd.DataFrame({
    "EBIT": EBIT,
    "NOPAT": NOPAT,
    "Invested Capital": Invested_Capital,
    "ROIC": ROIC
})

print(result)
```

```python
import yfinance as yf
import numpy as np

ticker = yf.Ticker("1812.T")

# Get data
net_income = ticker.income_stmt.loc["Net Income"]
equity = ticker.balance_sheet.loc["Stockholders Equity"]

roe = (net_income / equity).dropna().values[::-1]  # oldest → newest
print(roe )
years = np.arange(len(roe))

# Linear regression (slope)
slope = np.polyfit(years, roe, 1)[0]
latest_roe = roe[-1]

# Decision logic
if latest_roe < 0.05:
    result = "Cần cải thiện gấp"
elif slope > 0.005 and latest_roe >= 0.02:
    result = "Cải thiện tốt"
elif -0.005 <= slope <= 0.005:
    result = "Duy trì"
else:
    result = "Đang giảm"

print(f"Latest ROE: {latest_roe:.2%}")
print(f"ROE Slope: {slope:.2%}")
print(f"Assessment: {result}")
```

```python
import yfinance as yf
import plotly.graph_objects as go

ticker = yf.Ticker("9984.T")

# Financial statements (annual)
income = ticker.financials
balance = ticker.balance_sheet

# Dữ liệu
net_income = income.loc["Net Income"]
total_assets = balance.loc["Total Assets"]

# ROA
roa = net_income / total_assets

# Figure
fig = go.Figure()

# ROA (trục trái)
fig.add_trace(go.Scatter(
    x=roa.index,
    y=roa.values,
    mode="lines+markers",
    name="ROA",
    yaxis="y1"
))

# Net Income (trục phải)
fig.add_trace(go.Scatter(
    x=net_income.index,
    y=net_income.values,
    mode="lines+markers",
    name="Net Income",
    yaxis="y2"
))

# Total Assets (trục phải)
fig.add_trace(go.Scatter(
    x=total_assets.index,
    y=total_assets.values,
    mode="lines+markers",
    name="Total Assets",
    yaxis="y2"
))

# Layout
fig.update_layout(
    title="ROA, Net Income & Total Assets (1812.T)",
    template="plotly_white",

    xaxis=dict(title="Year"),

    yaxis=dict(
        title="ROA",
        tickformat=".2%",
        range=[0, 0.10]

    ),

    yaxis2=dict(
        title="JPY",
        overlaying="y",
        side="right",
        showgrid=False
    ),

    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig.show()
```

```python
import yfinance as yf
import numpy as np

# Load data
ticker = yf.Ticker("1812.T")

# Get historical Free Cash Flow (FCF)
cashflow = ticker.cashflow
fcf = cashflow.loc["Free Cash Flow"].dropna()
print(fcf)

# Assumptions
wacc = 0.08          # Discount rate
g = 0.02             # Terminal growth
years = len(fcf)

# Use latest FCF as base
fcf_base = fcf.iloc[0]

# Forecast FCF
fcf_forecast = [fcf_base * (1 + g) ** i for i in range(1, years + 1)]

# Discount FCF
discounted_fcf = [
    fcf_forecast[i] / (1 + wacc) ** (i + 1)
    for i in range(years)
]

# Terminal Value
terminal_value = fcf_forecast[-1] * (1 + g) / (wacc - g)
discounted_tv = terminal_value / (1 + wacc) ** years

# Enterprise Value
enterprise_value = sum(discounted_fcf) + discounted_tv

print(f"DCF Enterprise Value: {enterprise_value:,.0f} JPY")
```

```python
import yfinance as yf

# Load ticker
ticker = yf.Ticker("8801.T")

# Lấy Free Cash Flow lịch sử
cashflow = ticker.cashflow
fcf_series = cashflow.loc["Free Cash Flow"].dropna()

# Lấy 5 năm gần nhất (CF1 ... CFn)
cash_flows = fcf_series.iloc[:5].tolist()

# Discount rate
r = 0.08  # 8%

# DCF calculation (đúng công thức)
dcf_value = 0
for t, cf in enumerate(cash_flows, start=1):
    dcf_value += cf / (1 + r) ** t

print("Cash Flows (JPY):", cash_flows)
print(f"DCF Value (JPY): {dcf_value:,.0f}")
```

```python
# Terminal growth rate
g = 0.01  # 1% long-term growth (Japan reasonable)

fcf_n = cash_flows[-1]

terminal_value = (fcf_n * (1 + g)) / (r - g)

# Discount terminal value về hiện tại
terminal_value_pv = terminal_value / (1 + r) ** len(cash_flows)

print(f"Terminal Value (PV): {terminal_value_pv:,.0f}")
```

```python
enterprise_value_dcf = dcf_value + terminal_value_pv

print(f"DCF Enterprise Value: {enterprise_value_dcf:,.0f}")
```
