```python
import yfinance as yf

ticker = yf.Ticker("1812.T")  # Kajima Corporation

# ===== 1. BALANCE SHEET =====
bs = ticker.balance_sheet

# Lấy năm mới nhất
total_assets = bs.loc["Total Assets"].iloc[0]
total_liabilities = bs.loc["Total Liabilities Net Minority Interest"].iloc[0]

# Book Value of Assets (theo công thức bạn chọn)
book_value_assets = total_assets - total_liabilities


# ===== 2. MARKET VALUE =====
info = ticker.info

share_price = info.get("currentPrice")
shares_outstanding = info.get("sharesOutstanding")

market_value_assets = share_price * shares_outstanding


# ===== 3. ASSET REVALUATION RESERVE (PROXY) =====
asset_revaluation_reserve = market_value_assets - book_value_assets


# ===== 4. PRINT =====
print("=== Asset Revaluation Reserve (Market-based Proxy) ===")
print(f"Total Assets: {total_assets:,.0f}")
print(f"Total Liabilities: {total_liabilities:,.0f}")
print(f"Book Value of Assets: {book_value_assets:,.0f}")
print(f"Market Value (Equity): {market_value_assets:,.0f}")
print(f"Asset Revaluation Reserve (Proxy): {asset_revaluation_reserve:,.0f}")

```

```python
import yfinance as yf
import pandas as pd

ticker = yf.Ticker("1812.T")

# --- Income statement ---
income = ticker.financials
income_alt = ticker.income_stmt

# EBIT (fallback to Operating Income)
if "EBIT" in income.index:
    EBIT = income.loc["EBIT"]
else:
    EBIT = income_alt.loc["Operating Income"]

# Tax rate (single value, apply across years)
tax_rate = ticker.financials.loc["Tax Rate For Calcs"].iloc[0]

# --- Balance sheet ---
bs = ticker.balance_sheet

fixed_assets = bs.loc["Net PPE"]
current_assets = bs.loc["Current Assets"]
current_liabilities = bs.loc["Current Liabilities"]
goodwill = bs.loc["Goodwill And Other Intangible Assets"]

# --- Calculations ---
NOPAT = EBIT * (1 - tax_rate)

NWC = current_assets - current_liabilities

Invested_Capital = fixed_assets + NWC + goodwill

ROIC = (NOPAT / Invested_Capital).sort_index()

# --- Output ---
result = pd.DataFrame({
    "EBIT": EBIT,
    "NOPAT": NOPAT,
    "Invested Capital": Invested_Capital,
    "ROIC": ROIC
})

print(result)
```
