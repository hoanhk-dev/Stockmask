```python
import dspy

# 1. Khai báo signature: mô tả input/output
class SummarySignature(dspy.Signature):
    text = dspy.InputField()
    summary = dspy.OutputField(desc="short summary")

# 2. Xây module tóm tắt
class Summarizer(dspy.Module):
    def __init__(self):
        super().__init__()
        self.summarize = dspy.Predict(SummarySignature)

    def forward(self, text):
        return self.summarize(text=text).summary


# 3. Chạy
summarizer = Summarizer()
result = summarizer("So với các mô hình ngôn ngữ (LM) nguyên khối, mô hình module của DSPy cho phép một cộng đồng lớn cải thiện các kiến ​​trúc cấu thành, chiến lược thời gian suy luận và các thuật toán tối ưu hóa cho các chương trình LM theo cách mở và phân tán. Điều này mang lại cho người dùng DSPy nhiều quyền kiểm soát hơn, giúp họ lặp lại quy trình nhanh hơn nhiều và cho phép các chương trình của họ trở nên tốt hơn theo thời gian bằng cách áp dụng các thuật toán tối ưu hóa hoặc module mới nhất.")

print(result)
```
